###**CONVERTING DATATYPES INTO APPROPRIATE FIELDS**

```python
# Convert datatypes
# Converting datetime into correct format
df['date_added'] = pd.to_datetime(df['date_added'], errors = 'coerce')

# Converting release_year into Int type
df['release_year'] = df['release_year'].astype(int)
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

#Observations:
- The dataset contains Null values, Which are director column having 2634 Missing values, cast column having 825 values, Country having 831 missing values, date_added 98 missing values, rating 4, and duration 3.

#Using fillna

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['date_added'] = df['date_added'].fillna(0)
```

#Observation:
- The column like Director, cast, country, date_added Which have More than 100 Null values That's why we are using Fill Na Function For filling Null values.
- **NOTE :** If incase if we use Dropna The most of the dataset will be deleted so we are using Fill Na instead of Dropna.

#Using Dropna

```python
df.dropna(subset = ['rating', 'duration'], inplace = True)
```

#Observation:
- Using Dropna for rating & Duration Will not only Delete Null values it will aslo delete those entire entries.
- So Rating column has 4 Null values & Duration having 3 Null values, so the null values are less than 10 so we are using Dropna function it will delete only 7 entries from Dataset

###**RELATIONSHIP BETWEEN COUNTRY & CONTENT TYPE**

```python
top_countries = df['country'].value_counts().head(5).index
Filtered_data = df[df['country'].isin(top_countries)]

plt.figure(figsize = (8,4))
sns.countplot(data = Filtered_data, y = 'country', hue = 'type', palette = 'dark')
plt.title('TOP 5 COUNTRIES WITH CONTENT TYPE')
plt.xlabel('FREQUENCY')
plt.ylabel('COUNTRY')
plt.legend(title = 'CONTENT TYPE',fontsize = 'medium', frameon = False, ncol = 2, loc = 'lower right')
plt.xticks(rotation = 0)
plt.yticks(rotation = 0)
plt.show()
```

#Observations:
- Most movies and TV shows on netflix come from the U.S. Movies are more than TV shows.
- A large number of shows and movies don't have a country name. In that Slightly more movies than TV shows.
- Indian movies are highly present on Netflix, Mostly movies, very few Tv shows.
- UK has a balanced amount of movies and Tv shows, Through Tv shows are a bit more.
- Japan has More Tv shows than movies, Lower overall content than the others.

###**BEST TIME TO LAUNCH TV SHOWS**

```python
df['month_name'] = pd.to_datetime (df['date_added']).dt.month_name().fillna('Unknown')
plt.figure(figsize = (14,6))
sns.countplot(x = df['month_name'], data = df[df['type'] == 'TV Show'], palette = 'muted')
plt.title('BEST TIME TO LAUNCH TV SHOWS', fontsize = 14)
plt.xlabel('MONTHS', fontsize = 12)
plt.ylabel('FREQUENCY', fontsize = 12)
plt.xticks(rotation = 0)
plt.show()
```
